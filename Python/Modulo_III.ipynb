## Capítulo 1: ndarrays <a class="anchor" id="first-bullet"></a>

Numpy nos permite crear ndarrays que son como listas pero más potentes, especialmente para cálculo matricial

| Función | Descripción |
| --- | --- |
| array | Convertir lista en ndarray |
| arange | Como range() pero devuelve un ndarray|
| linspace | Devuelve un ndarray con (start, stop, steps) |
| zeros | Produce un ndarray de ceros con el tamaño especificado |
|zeros_like | Produce un ndarray de ceros con un tamaño igual a otro array |
| ones | Produce un ndarray de unos con el tamaño especificado |
|ones_like | Produce un ndarray de unos con un tamaño igual a otro array |
| full | Produce un ndarray con el valor especificado como relleno |
| fulls_like| Produce un ndarray con el valor especificado como relleno y un tamaño igual a otro array | 

```python
lista = [0,1,2,3,4,"cinco"]

array = np.array(lista)

print(array)
```

```python
arr1 = np.arange(6) 

print(arr1, "\n")

arr2 = np.linspace(0,50,11)

print(arr2, "\n")

zeros = np.zeros_like(arr2)

print(zeros, "\n")

cincos = np.full(10, fill_value = 5)

print(cincos)
```

También nos permite crear arrays multidimensionales (matrices, tensores etc.) 

```python
matriz = np.array([[1,2,3,4],[5,6,7,8]])

print(matriz, "\n")

print(matriz[0])   # primera fila
print(matriz[1,3]) # segunda fila, cuarto elemento
print(matriz[-1,-2]) # última fila, penúltimo elemento
```

También podemos hacer operaciones siempre y cuando las matrices tengan las mismas dimensiones

```python
unos = np.ones_like(matriz)

print(matriz - unos, "\n")
print(matriz + unos, "\n")
print(matriz * 4, "\n")
print(matriz ** 3, "\n")
print(1 / matriz, "\n")
print(matriz < 5, "\n")
```

## Capítulo 2: Slicing <a class="anchor" id="second-bullet"></a>

Para acceder a los elementos usamos el slicing igual que con las listas

```python
Image("fig_slice.png")
```

```python
arr = np.arange(9).reshape([3,3])

print(arr,"\n")

print(arr[:2,1:],"\n")

print(arr[2,:],"\n")
```

```python
arr = np.arange(27).reshape([3,3,3])

print("full array: \n\n",arr,"\n")

print("[2,1,1] : \n\n",arr[2,1,1],"\n")

print("[1:,2,:] : \n\n",arr[1:,2,:],"\n")

print("[:,:2, :-1] : \n\n",arr[:,:2, :-1],"\n")

print("[2,::-1, ::-1] : \n\n",arr[2,::-1, ::-1],"\n")
```

También se puede especificar que filas o columnas queremos

```python
print("full array: \n\n",arr,"\n")

print("[[2,1,2]]: \n\n",arr[[2,1,2]],"\n")

print("[1,[2,1,2],::-1]: \n\n",arr[1,[2,1,2],::-1],"\n")
```

## Capítulo 4: Números Pseudoaleatorios <a class="anchor" id="fourth-bullet"></a>

Numpy tiene integradas un gran número de distribuciones a través del submódulo **random** 

| Método | Descripción |
| --- | --- |
|permutation | Devuelve una permutación aleatoria de un ndarray |
| shuffle | Baraja el ndarray |
| uniform | Muestras de una distribución uniforme |
| randint | Muestras de números enteros entre *low* y *high* |
| standard normal | Muestras de una distribución estandar |
| binomial | Muestras de una distribución binomial |
| normal | Muestras de una distribución normal |
| beta | Muestras de una distribucion beta |

```python
arr = np.arange(10)

print(np.random.permutation(arr))
print(np.random.shuffle(arr))
print(arr)
```

```python
integers = np.random.randint(2,6, size = 4)
normal = np.random.normal(loc = 5, scale = 1, size = [3,4])

print(integers, "\n")

print(normal)
```

## Capítulo 5: Álgebra Lineal <a class="anchor" id="fifth-bullet"></a>

Numpy también viene integrado con una serie de funciones ideal para cálculos de Álgebra lineal y resolver sistemas de ecuaciones.

El operador **@** ésta reservado para el producto entre vectores mientras que el operador **\*** multiplaca cada elemento de forma paralela 

```python
x = np.array([3, 4, 1])
y = np.array([2, -3, 4])


print("Función np.dot: ", np.dot(x, y), "\n")
print("Operador @: ", x @ y, "\n")
print("Operador *: ", x * y)
```

Lo mismo se aplica para matrices

```python
A = np.array([[1, -2, 1], [2, -4, 5], [2, 5,  1]])

B = np.array([[3, 4, 2], [1, -2, 0], [5, 1, -4]])

print("Función np.matmul: \n", np.matmul(A, B), "\n")
print("Operador @: \n", A @ B, "\n")
print("Operador *: \n", A * B)
```

Y para productos entre matrices y vectores

```python
print("Función np.dot: \n", np.dot(A, x), "\n")
print("Operador @: \n", A @ x, "\n")
print("Operador *: \n", A * x)
```

Dentro del submódulo **np.linalg** podemos encontrar un gran número de funciones y operaciones comunes en Álgebra Lineal:

| Función | Descripción |
| --- | --- |
| inv | Invierte una matriz |
| det | Calcula el determinante de una matriz |
| cholesky | Descomposición de Cholesky de una matriz |
| svd | Descomposición de valores singulares de una matriz |
| eig | Valores propios y vectores propios de una matriz | 
| solve | Resuelva un sistema lineal de ecuaciones |

Por ejemplo, las ecuaciones normales de mínimos cuadrados que resuelven un problema de regresión lineal son:

$$
\beta = (X^{T} X)^{-1} X^{T} y
$$

```python
beta = np.linalg.inv(A.T @ A) @ A.T @ x

print("beta = ", beta)
```
